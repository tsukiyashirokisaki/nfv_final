```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
def read_data(root):
    times = []
    header = []    
    for name in sorted(os.listdir(root)):
        with open(os.path.join(root,name),"r") as f:
            times.append(f.read().split("\n")[:10])
            header.append(name)
    times = np.array(times).astype("float")
    means = np.mean(times,axis=1)
    stds = np.std(times,axis=1)
    plt.errorbar(range(9),means,stds,marker='o',capsize = 3)
    my_xticks = ['baseline']+["%d core"%(i) for i in range(1,9)]
    plt.xticks(range(9), my_xticks,fontsize=10)
    plt.xlabel("Cores",fontsize=16)
    plt.ylabel("Running Time (s)",fontsize=16)
    plt.tight_layout()
    plt.savefig("plot/%s.png"%(root.replace("/","_")),dpi=600)
    print(means)
#     print(means[0]/means[-1])
# read_data("frag_inv_dump")
read_data("wc/frag")
```
